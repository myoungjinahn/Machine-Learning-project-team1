- 초미세먼지 농도 15 이하 => 좋음
- 초미세먼지 농도 15 초과 => 나쁨

```python


# =========================
# 1. 엑셀 파일 불러오기
# =========================

df = pd.read_excel("23년부터 25년 3년 가을 초미세먼지 데이터 완성본.xlsx")
# =========================
# 라이브러리 불러오기
# =========================

import pandas as pd
import numpy as np

# =========================
# 파일 이름 입력
# =========================

file_path = "23년부터 25년 3년 가을 초미세먼지 데이터 완성본.xlsx"

# =========================
# 시트 목록 확인
# =========================

excel_file = pd.ExcelFile(file_path)

print(excel_file.sheet_names)

# =========================
# 사용할 시트
# =========================

sheet_list = [
    '25.11', '25.10', '25.09',
    '24.11', '24.10', '24.09',
    '23.11', '23.10', '23.09'
]

all_data = []

# =========================
# 데이터 추출
# =========================

for sheet in sheet_list:

    # 시트 읽기
    df = pd.read_excel(
        file_path,
        sheet_name=sheet,
        header=None
    )

    # 서울시 평균 행 찾기
    seoul_row = df[df[0] == '서울시 평균']

    # 없으면 건너뛰기
    if seoul_row.empty:
        continue

    seoul_row = seoul_row.iloc[0]

    # 연도 / 월 추출
    year = "20" + sheet.split('.')[0]
    month = sheet.split('.')[1]

    # 일별 데이터 추출
    for day in range(1, 32):

        try:
            value = seoul_row[day + 1]

            # 결측치 제거
            if pd.isna(value):
                continue

            # 날짜 생성
            date = f"{year}-{month}-{day:02d}"

            all_data.append({
                "date": date,
                "pm25": value
            })

        except:
            continue

# =========================
# 데이터프레임 생성
# =========================

result_df = pd.DataFrame(all_data)

# 날짜 타입 변환
result_df["date"] = pd.to_datetime(result_df["date"])

# =========================
# 라벨링 함수 (좋음 = 1, 나쁨 = 0)
# =========================

def pm25_label(value):

    if value <= 15:
        return 1   # 좋음
    else:
        return 0   # 나쁨

# =========================
# 라벨 컬럼 생성
# =========================

result_df["label"] = result_df["pm25"].apply(pm25_label)

# =========================
# 결과 확인
# =========================

print(result_df.head())

# =========================
# CSV 저장
# =========================

result_df.to_csv(
    "pm25_labeled_data.csv",
    index=False,
    encoding="utf-8-sig"
)

print("라벨링 완료")
```

# 온도 분류 및 라벨링

분류
- 최고기온 11°C ~ 18°C -> 쾌적(1)
- 그 이외의 기온 -> 불쾌적(0)

2. 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
import os

# 업로드된 파일 이름 확인
print(os.listdir())
```

3. 파일 이름 입력

```python
file_path = "23-25년 09-11월 온도통합데이터 완성본.xlsx"
```

4. 엑셀 읽기

```python
df = pd.read_excel(
    file_path,
    sheet_name="통합데이터",
    header=None
)
```

5. 데이터 저장 리스트

```python
all_data = []
```

6. 2023 데이터 추출

```python
for i in range(10, len(df)):

    try:
        date = df.iloc[i, 0]
        temp = df.iloc[i, 2]

        if pd.isna(date) or pd.isna(temp):
            continue

        all_data.append({
            "date": date,
            "temp": temp
        })

    except:
        continue
```

7. 2024 데이터 추출

```python
for i in range(10, len(df)):

    try:
        date = df.iloc[i, 5]
        temp = df.iloc[i, 7]

        if pd.isna(date) or pd.isna(temp):
            continue

        all_data.append({
            "date": date,
            "temp": temp
        })

    except:
        continue
```

8. 2025 데이터 추출

```python
for i in range(10, len(df)):

    try:
        date = df.iloc[i, 10]
        temp = df.iloc[i, 12]

        if pd.isna(date) or pd.isna(temp):
            continue

        all_data.append({
            "date": date,
            "temp": temp
        })

    except:
        continue
```

9. 데이터프레임 생성

```python
result_df = pd.DataFrame(all_data)

# 날짜 타입 변환
result_df["date"] = pd.to_datetime(result_df["date"])
```

10. 라벨링 함수

```python
def temp_label(value):

    if 11 <= value <= 18 :
        return 1   # 쾌적
    else:
        return 0   # 불쾌적
```

11. 라벨 생성

```python
result_df["label"] = result_df["temp"].apply(temp_label)
```

12. 결과 확인

```python
print(result_df.head())

print("\n데이터 개수:")
print(result_df.shape)
```

13. CSV파일 저장

```python
result_df.to_csv(
    "temperature_labeled_data.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\n라벨링 완료")
```
